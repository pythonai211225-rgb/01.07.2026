```python
import numpy as np

# Data
x = np.arange(0, 5, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
y_exp = np.exp(x)  # Example additional function

# Create figure and axes for subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

# 0,0 0,1
# 1,0 1,1

# Sine plot
ax[0][0].plot(x, y_sin, label='sin(x)', color='blue')
ax[0, 0].set_title('Sine Function')
ax[0, 0].set_xlabel('X Values')
ax[0, 0].set_ylabel('sin(x)')
ax[0, 0].legend()
ax[0, 0].grid(True)

# Cosine plot
ax[0, 1].plot(x, y_cos, label='cos(x)', color='red')
ax[0, 1].set_title('Cosine Function')
ax[0, 1].set_xlabel('X Values')
ax[0, 1].set_ylabel('cos(x)')
ax[0, 1].legend()
ax[0, 1].grid(True)

# Tangent plot
ax[1, 0].plot(x, y_tan, label='tan(x)', color='green')
ax[1, 0].set_title('Tangent Function')
ax[1, 0].set_xlabel('X Values')
ax[1, 0].set_ylabel('tan(x)')
ax[1, 0].legend()
ax[1, 0].grid(True)

# Exponential plot
# e = 2.718
ax[1, 1].plot(x, y_exp, label='exp(x)', color='purple')
ax[1, 1].set_title('Exponential Function')
ax[1, 1].set_xlabel('X Values')
ax[1, 1].set_ylabel('exp(x)')
ax[1, 1].legend()
ax[1, 1].grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot with all the mentioned features
# First line: customized color, line width (lw), line style (ls), marker, and marker size
ax.plot(x, y, label='Sine Wave', color='#1FEBDD', lw=2, ls='--', marker='o', markersize=6)

# Second line: another set of customization
ax.plot(x, y2, label='Cosine Wave', color='red', lw=3, linestyle='-', marker='x', markersize=8)

# Set title, labels, and legend
ax.set_title('Customized Chart', fontsize=14)
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)

# Add legend
ax.legend()

# Add grid
ax.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Define the data
movies = ['Sci-Fi', 'Comedy', 'Action', 'Drama']
votes = [45, 25, 60, 15]

# 2. Define a list of colors (one for each bar)
bar_colors = ['skyblue', 'gold', 'lightgreen', 'tomato']

# 3. Create the bar chart and apply the colors
plt.bar(movies, votes, color=bar_colors, edgecolor='black')

# 4. Add labels and make it look clean
plt.ylabel('Number of Votes')
plt.title('Favorite Movie Genres')

plt.tight_layout() # Keeps everything fitting nicely!
plt.show()
```

```python

# 1. Define the data (Example: Video game sales)
games = ['Game A', 'Game B', 'Game C']
digital_sales = [40, 55, 30]  # This will be the bottom color
physical_sales = [25, 20, 45] # This will be the top color

# 2. Draw the bottom bars (Blue)
plt.bar(games, digital_sales, color='skyblue', label='Digital')

# 3. Draw the top bars (Orange) 
# The magic trick is using: bottom=digital_sales
plt.bar(games, physical_sales, bottom=digital_sales, color='orange', label='Physical')

# 4. Clean up the graph
plt.ylabel('Units Sold (Thousands)')
plt.title('Game Sales Breakdown')
plt.legend() # Shows which color is which

plt.tight_layout()
plt.show()
```

```python

# 1. Define the data
categories = ['Apples', 'Bananas', 'Cherries']
left_values = [10, 15, 7]
right_values = [12, 8, 14]

# 2. Set up the bar positions on the X-axis
x = np.arange(len(categories)) 
width = 0.35  # How wide each half-bar will be

# 3. Draw the left half and right half so they touch perfectly
# Subtracting/adding 'width/2' shifts them so they meet in the middle
plt.bar(x - width/2, left_values, width=width, color='skyblue', label='Left Side')
plt.bar(x + width/2, right_values, width=width, color='salmon', label='Right Side')

# 4. Fix the X-axis labels so they sit right under the split
plt.xticks(x, categories)

# 5. Clean up the graph
plt.ylabel('Values')
plt.title('Vertically Split Bars')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

slices = [40, 30, 20, 10]
activities = ['Gaming', 'Sleeping', 'School', 'Eating']
pie_colors = ['skyblue', 'gold', 'lightgreen', 'tomato']

plt.pie(
    slices, 
    labels=activities, 
    colors=pie_colors,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5} # <-- Use this instead!
)

plt.tight_layout()
plt.show()
```
