```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 9 + 1)

# 1 - long
y2 = []
for num in x:
    y2.append(num * 2 + 1)
print(y2)    

# 2 - short
y = x * 2 + 1

print(x)
print(y)
```

```python
plt.plot(x, y)
y1 = y + 0.5
plt.plot(x, y1)
plt.title("Hours vs Production")
plt.xlabel('Hours')
plt.ylabel('Production')
plt.legend(['Production', 'Product + 0.5'])
plt.xlim(0, 6)
plt.ylim(0, 12.5)
plt.savefig('line_chart1.jpg')
plt.show()

# make x points from 5 to 15 , make 20 points
# y = 3 * x - 1
# draw the graph using plot
```
