```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6.4, 4.8))  # 640 ==> 6.4, 480 => 4.8

#            left=0 bottom=0 width=% height=%
axes1 = fig.add_axes([0, 0, 1, 1])
x = np.arange(0, 10)
y = 2 * x
axes1.set_xlim(0, 6)
axes1.set_ylim(0, 12.5)
axes1.set_xlabel('Hours')
axes1.set_ylabel('Production')
axes1.set_title('Graph 1')
axes1.plot(x, y)

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.3])
a = np.arange(0, 10)
b = a ** 4
axes2.set_title('')
axes2.set_title('Zoomed In [Graph 2]')
axes2.plot(a, b)

axes3 = fig.add_axes([1.2, 0, 1, 1])
axes3.set_title('Graph 3')
a = np.arange(0, 10)
b = a ** 4
axes3.plot(a, b)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

###################### # solution for sinq cos of matzeget 23 final exercise

fig = plt.figure(figsize=(7,2)) 

x = np.arange(0, 5, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

ax1 = fig.add_axes([0, 0, 1, 1])  # Main axes
ax1.plot(x, y_sin, label='sin(x)', color='blue')
ax1.set_title('Sine Function')
ax1.set_xlabel('X Values')
ax1.set_ylabel('sin(x)')
ax1.legend()
ax1.grid(True)  # Optional: Add gridlines

# Step 6: Plot the cos() function on the second subplot
ax2 = fig.add_axes([1.2, 0, 0.5, 1])  # Smaller, inset axes
ax2.plot(x, y_cos, label='cos(x)', color='orange')
ax2.set_title('Cosine Function')
ax2.set_xlabel('X Values')
ax2.set_ylabel('cos(x)')
ax2.legend()
ax2.grid(False) 

plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# fig = plt.figure(figsize=(7,2)) 
# ax1 = fig.add_axes([0, 0, 1, 1])  # Main axes
# ax1.plot(x, y_sin, label='sin(x)', color='blue')
# ax1.set_title('Sine Function')

# Create figure and axes for subplots
# subplot == table!!
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))  # 2 cols, 1 row
# Sine plot
ax[0].plot(x, y_sin, label='sin(x)', color='blue')
ax[0].set_title('Sine Function')
ax[0].set_xlabel('X Values')
ax[0].set_ylabel('sin(x)')
ax[0].legend()
ax[0].grid(True)

# Cosine plot
ax[1].plot(x, y_cos, label='cos(x)', color='red')
ax[1].set_title('Cosine Function')
ax[1].set_xlabel('X Values')
ax[1].set_ylabel('cos(x)')
ax[1].legend()
ax[1].grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
