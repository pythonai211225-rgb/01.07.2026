```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 9 + 1)

# 1 - long
y2 = []
for num in x:
    y2.append(num * 2 + 1)
print(y2)    

# 2 - short
y = x * 2 + 1

print(x)
print(y)
```

```python
plt.plot(x, y)
y1 = y + 0.5
plt.plot(x, y1)
plt.title("Hours vs Production")
plt.xlabel('Hours')
plt.ylabel('Production')
plt.legend(['Production', 'Product + 0.5'])
plt.xlim(0, 6)
plt.ylim(0, 12.5)
plt.savefig('line_chart1.jpg')
plt.show()

# make x points from 5 to 15 , make 20 points
# y = 3 * x - 1
# draw the graph using plot
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6.4, 4.8))  # 640 ==> 6.4, 480 => 4.8

#            left=0 bottom=0 width=% height=%
axes1 = fig.add_axes([0, 0, 1, 1])
x = np.arange(0, 10)
y = 2 * x
axes1.set_xlim(0, 6)
axes1.set_ylim(0, 12.5)
axes1.set_xlabel('Hours')
axes1.set_ylabel('Production')
axes1.set_title('Graph 1')
axes1.plot(x, y)

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.3])
a = np.arange(0, 10)
b = a ** 4
axes2.set_title('Zoomed In [Graph 2]')
axes2.plot(a, b)

axes3 = fig.add_axes([1.2, 0, 1, 1])
axes3.set_title('Graph 3')
a = np.arange(0, 10)
b = a ** 4
axes3.plot(a, b)

plt.show()
```
