```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')

# plt.scatter(data=df, x='salary', y='sales')

sns.scatterplot(data=df, x='salary', y='sales')
plt.show()

display(df)
```

```python
plt.figure(figsize=(8,4))
# sns.scatterplot(data=df, x='salary', y='sales', hue='sales', palette='Spectral')

sns.scatterplot(data=df, x='salary', y='sales', hue='work experience', palette='Spectral')

plt.tight_layout()
plt.legend(loc=(0.1, 0.5))

plt.show()
```

```python
plt.figure(figsize=(8,4)) 

df = pd.read_csv("dm_office_sales.csv")

#sns.scatterplot(data=df, x='salary', y='sales',hue="salary")
#sns.scatterplot(data=df, x='salary', y='sales',hue="work experience", palette='dark')
sns.scatterplot(data=df, x='salary', y='sales',s=120, hue="training level", palette='Spectral', alpha=0.7, marker='o')

plt.tight_layout()
plt.legend(loc=(1.1, 0))
plt.show()
```
